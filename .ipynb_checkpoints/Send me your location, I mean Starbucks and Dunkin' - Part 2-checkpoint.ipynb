## Dunkin' really loves transit station, car repair store and gas station

### The number of each type of merchant should follow Multinomial + Dirichlet distribution

If we evaluate preference using probability, then the number of each type is close to $N_{AllType}*P_{ThisType}$. Supposedly, the number of each type should follow [Multinomial distribution](https://en.wikipedia.org/wiki/Multinomial_distribution) with a prior [Dirichlet distribution](https://en.wikipedia.org/wiki/Dirichlet_distribution).  
$$ (N_{Type_A}, N_{Type_B},..., N_{Type_Z}) \sim Multinomial(N_{Type_{All}}, p_{Type_A}, p_{Type_B}... , p_{Type_Z}) $$
$$ (p_{Type_A}, p_{Type_B}, ... , p_{Type_Z}) \sim Dirichlet(\alpha_{Type_A}, \alpha_{Type_B}, ... , \alpha_{Type_Z}) $$    
The plots below describe the distribution of probability for each type of merchant. Most of them seem follow beta distribution. However, some plots (e.g. ATM, Bank, Cafe, gym, laundry and others) show obvious double peaks, which means inconsistent preferences might be possible within Starbucks and Dunkin'.

```python
nearby_prob = pd.DataFrame(nearby[nearby_type].values/ nearby[nearby_type].sum(axis=1).values.reshape(-1,1), columns = nearby_type)
f, axarr = plt.subplots(4, 5, figsize = (20,20))
for i in range(4):
    for j in range(5):
        if (i * 5 + j) == len(nearby_type):
            break
        s = sns.distplot(nearby_prob[nearby_type[i * 5 + j]][(nearby['type']=='starbucks')], color = 'green', ax = axarr[i,j], bins=10)
        s = sns.distplot(nearby_prob[nearby_type[i * 5 + j]][(nearby['type']=='dunkin')], color = 'red',ax = s, bins=10)
        s.set(xlim=(0,None))
plt.show()
```

To evaluate preference, I fit the prior distribution for Starbucks and Dunkin' seperately, and update posterior distribution as a part of [Emperical Bayes method](https://en.wikipedia.org/wiki/Empirical_Bayes_method). After that, calibrated [Postorior Predictive P-value](http://www.stat.columbia.edu/~gelman/research/published/A6n41.pdf) was used to find out specific stores which was obviously different from the group. It is alternative [chi-square test for independence](https://en.wikipedia.org/wiki/Chi-squared_test) in Bayesian Statistics.

```python
#Try to combine all regions
slice_starbucks = (nearby['type'] == 'starbucks')
slice_dunkin = (nearby['type'] == 'dunkin')
alpha_prior_starbucks = fit_mvpolya(nearby[nearby_type][slice_starbucks].values)['params']
alpha_prior_dunkin = fit_mvpolya(nearby[nearby_type][slice_dunkin].values)['params']
alpha_whole_starbucks = alpha_prior_starbucks + nearby[nearby_type][slice_starbucks].values.sum(axis=0)
alpha_whole_dunkin = alpha_prior_dunkin + nearby[nearby_type][slice_dunkin].values.sum(axis=0)
```

The plot of probability of each merchants from Bayesian model is pretty similar to the plot of the average number of each merchants in the previous part.    

However, Dunkin' has a much stronger preference of transit station, car repair store, gas station and church than Starbucks. Besides, Dunkin' is more likely to be in the area with laundry store, liquor store, park and school. In contrast, Starbucks has a relatively stronger preference with bank, beauty salon and gym than Dunkin'. It basically comforms to the different markets they focuses on (i.e, commercial and bussiness districts for Starbucks; resident districts and places related to commuting, like transit station and high way, for Dunkin').

```python
#Sampling from dirichlet distribution
def dirichlet_rvs(alpha):
    y = np.array([])
    for a in alpha:
        y = np.append(y, np.random.gamma(a))
    return y / sum(y)

#Compare each distribution: Return estimation and CI of scaled difference (i.e. (p_x-p_y) / (p_x+p_y))
def comparison(alpha1, alpha2):
    diff = np.array([])
    for i in range(5000):
        p1 = dirichlet_rvs(alpha1)
        p2 = dirichlet_rvs(alpha2)
        diff = np.append(diff, (p1 - p2)/(p1 + p2))
    diff = diff.reshape(5000,-1).T
    estimator = np.percentile(diff, 50, axis = 1)
    ci_lower = np.percentile(diff, 2.5, axis = 1)
    ci_upper = np.percentile(diff, 97.5, axis = 1)
    return {'estimator': estimator, 'ci_lower': ci_lower, 'ci_upper': ci_upper}
```

```python
nearby_sum_bayes = pd.DataFrame({'merchant': np.append(nearby_type, nearby_type), 
                                 'prob': np.append(alpha_whole_dunkin/sum(alpha_whole_dunkin), alpha_whole_starbucks/sum(alpha_whole_starbucks)),
                                 'type': np.array(['dunkin']*len(nearby_type) + ['starbucks']*len(nearby_type))})

ci = comparison(alpha_whole_starbucks, alpha_whole_dunkin)
ci['ci_lower'] = ci['ci_lower'][:-1]
ci['ci_upper'] = ci['ci_upper'][:-1]
ci['estimator'] = ci['estimator'][:-1]
ci['merchant'] = nearby_type[:-1].tolist()
ci['starbucks'] = nearby_sum_bayes[nearby_sum_bayes['type'] == 'starbucks'].prob.values[:-1].tolist()
ci['dunkin'] = nearby_sum_bayes[nearby_sum_bayes['type'] == 'dunkin'].prob.values[:-1].tolist()
```

```python
from bokeh.models import ColumnDataSource, HoverTool, Span
from bokeh.plotting import figure
from bokeh.transform import dodge
from bokeh.core.properties import value
from bokeh.layouts import gridplot
from bokeh.io import output_notebook, show

output_notebook()

source = ColumnDataSource(ci)

hover_left = HoverTool(tooltips=[
                                ("Merchant", "@merchant"),
                                ("P_Starbucks", "@starbucks{0,0.000}"),
                                ("P_Dunkin", "@dunkin{0,0.000}"),
                                ])
hover_right = HoverTool(tooltips=[
                                ("Merchant", "@merchant"),
                                ("Credible Interval", "[@ci_lower{0,0.000}, @ci_upper{0,0.000}]"),
                                ("Estimator", "@estimator{0,0.000}"),
                                ])
TOOLS = "hover, tap"
left_bar = figure(title="Probability of each merchant", plot_width=400, plot_height=300, tools='tap', y_range= nearby_type[:-1].tolist())
left_bar.hbar(y = dodge('merchant', -0.25, range=left_bar.y_range), right = 'starbucks', height=0.2, source=source, color='#008000', legend=value("Starbucks"))
left_bar.hbar(y = dodge('merchant', 0.25, range=left_bar.y_range), right = 'dunkin', height=0.2, source=source, color='#FF0000', legend=value("Dunkin\'"))
left_bar.add_tools(hover_left)
left_bar.legend.label_text_font_size = '6pt'
left_bar.legend.location = 'center_right'
left_bar.legend.label_text_font_style = 'bold'

zero = Span(location=0,
            dimension='height', line_color='red',
            line_dash='dashed', line_width=3)
right_ci = figure(tools='tap', plot_width=300, plot_height=300, title="(P_S - P_D) / (P_S + P_D)", y_range= nearby_type[:-1].tolist())
right_ci.yaxis.visible = False
right_ci.hbar(y="merchant", left='ci_lower', right='ci_upper', height=0.1, source=source, line_cap = 'butt')
right_ci.diamond(y="merchant", x = 'estimator', source=source, size = 7)
right_ci.add_layout(zero)
right_ci.add_tools(hover_right)

p = gridplot([[left_bar, right_ci]])

show(p)
```

## Interestingly, major commercial areas are the most consistent regions for Dunkin' but the least for Starbucks

More than half of Starbucks stores in major commercial areas and Allston are different from the average level. In contrast, the most inconsistent regions for Dunkin' are the places outside of major commercial areas.

```python
import pymc3 as pm
#Discrepancy measurement for Postorior predictive p-value
def discrepancy(y, p):
    y_expectation = p * sum(y)
    return sum(np.array([(y1 - y2)**2 for y1, y2 in zip(y, y_expectation)]))
    
#Gibbs samppling for Dirichlet-Multinomial Distribution
def MCMC(iterations, burn, alpha, n):
    p = dirichlet_rvs(alpha)
    y = np.random.multinomial(n, p)
    p_simulation = []
    y_simulation = []
    for i in range(iterations+burn):
        alpha = alpha + y
        p = dirichlet_rvs(alpha)
        y = np.random.multinomial(n, p)
        if i >= burn:
            p_simulation.append(p)
            y_simulation.append(y)   
    return (p_simulation, y_simulation)

#Postorior predictive p-value
def ppp(observation, alpha, iteration = 500, burn = 100):
    p_simulation, y_replicate = MCMC(iteration, burn ,alpha , sum(observation))
    d_obs = []
    d_sim = []
    for p, y in zip(p_simulation, y_replicate):
        d_obs.append(discrepancy(observation, p))
        d_sim.append(discrepancy(y, p))
    return np.mean(np.array(d_sim) >= np.array(d_obs))

#Posterior Calibarated Postorior predictive p-value
def post_cppp(observation, alpha, iteration = 500, burn = 100):
    ppp_obs = np.array([ppp(observation, alpha)])
    
    ppp_sim = []
    p_simulation, y_replicate = MCMC(iteration, burn ,alpha , sum(observation))
    i = 1
    for y in y_replicate:
        ppp_sim.append(ppp(y,alpha))
        if i%100 == 0:
            print("{}% is completed".format(i / 10.0))
    
    return np.mean(np.array([ppp_obs]) >= np.array(ppp_sim))
```

```python
alpha_individual = nearby[nearby_type].values.astype(float)
alpha_individual[slice_starbucks] = alpha_prior_starbucks + alpha_individual[slice_starbucks]
alpha_individual[slice_dunkin] = alpha_prior_dunkin + alpha_individual[slice_dunkin]
results_all = []
for i in range(len(nearby)):
    print("Work on from row {}".format(i))
    results_all.append(post_cppp(nearby[nearby_type].values[i], alpha_individual[i]))
    print("{}% is done.".format((i+1)/len(nearby)*100))
    with open('./data/results_all.pkl','wb') as f:
        pickle.dump(results_all, f)
```

## Commercial areas are featured with bank and cafe while residential areas are featured with gas stations, car repair stores and transit station

For either Dunkin' or Starbucks, there are obviously less gas stations and car repair stores in major commercial areas, comparing to each of their average levels. In contrast, stores in residential areas have way less banks but more gas stations, car repair stores and transit station. Besides, the number of cafe is much smaller in the regions like Charlestown, Dorchester, Jamaica, Chelsea.

```python
def alpha_region(datasets, region_list = np.unique(nearby['region']), nearby_type = nearby_type):
    alpha = {}
    for region in region_list:
        datasets_this_region = datasets[nearby_type][datasets['region'] == region]
        alpha[region] = fit_mvpolya(datasets_this_region.values)['params'] + datasets_this_region.sum(axis=0).values
    return alpha
```

```python
alpha = {'dunkin': alpha_region(nearby[slice_dunkin]),
         'starbucks':alpha_region(nearby[slice_starbucks])}
l_s = [k for k in alpha['starbucks'].keys() if any(np.isnan(alpha['starbucks'][k]))]
l_d = [k for k in alpha['dunkin'].keys() if any(np.isnan(alpha['dunkin'][k]))]
for l in l_s:
    del alpha['starbucks'][l]
for l in l_d:
    del alpha['dunkin'][l]
prob = {'dunkin': {k:alpha['dunkin'][k]/sum(alpha['dunkin'][k]) for k in alpha['dunkin'].keys()},
        'starbucks':{k:alpha['starbucks'][k]/sum(alpha['starbucks'][k]) for k in alpha['starbucks'].keys()}}
```

```python
preference_type = ['dunkin'] * len(alpha['dunkin'].values()) + ['starbucks'] * len(alpha['starbucks'].values())
preference_region = [key for key in alpha['dunkin'].keys()] + [key for key in alpha['starbucks'].keys()]
preference_description = ["{},{}".format(t,r) for t, r in zip(preference_type, preference_region)]

#Write program to for loop and save it
compare = {}
for i, r in enumerate(preference_description):
    if preference_type[i] == 'starbucks':
        alpha_whole = alpha_whole_starbucks
    else:
        alpha_whole = alpha_whole_dunkin
    compare[r] = comparison(alpha[preference_type[i]][preference_region[i]],
                            alpha_whole)
```

# Documentation

**Name**: dirichlet_rvs

**Capability**: 
It is used to draw random samples from dirichlet distribution using Gamma distribution. To see more, check [here](https://stats.stackexchange.com/questions/69210/drawing-from-dirichlet-distribution). It solves the problem within numpy.random.dirichlet function when one of alpha is much smaller the others.

**Input**:

|variable|Type|Description|

**Name**: comparison

**Capability**: 
Get estimator and 95% credible interval of $(p_x-p_y) / (p_x+p_y)$ from simulation.

**Input**:

|variable|Type|Description|
|:------|:---|:--------|
|alpha1|k-dimension array of floats|Parameter of the distribution x|
|alpha2|k-dimension array of floats|Parameter of the distribution y|

**Output**: A dict contains the following returns.

|variable|Type|Description|
|:------|:---|:--------|
|estimator|array of floats|Estimation of $(p_x-p_y) / (p_x+p_y)$ for each merchant|
|ci_lower|array of floats|Lower boundary of 95% credible interval|
|ci_upper|array of floats|Upper boundary of 95% credible interval|

**Name**: discrepancy

**Input**:

|variable|Type|Description|
|:------|:---|:--------|
|y|array of ints|The actual number of observations for each type of merchants|
|p|array of floats|The probability of each type of merchants|

**Output**: A float used for discrepancy measurement

**Name**: MCMC

**Capability**: 
[Gibbs samppling](https://en.wikipedia.org/wiki/Gibbs_sampling) for Dirichlet-Multinomial Distribution.

**Input**:

|variable|Type|Description|
|:------|:---|:--------|
|iterations|int|The number of random draws from MCMC|
|burn|int|The number of loops to warm-up for MCMC|
|alpha|k-dimension array of floats|Parameter of the distribution (k dimension for sample of dimension k)|
|n|int|The number of all kinds of merchants in the neighborhood|

**Output**: $j*k$-dimension array of floats: j draws of random samples from k-dimension dirichlet distribution

**Name**: post_cppp

**Capability**: 
It is used to calibarate postorior predictive p-value.

**Input**:

|variable|Type|Description|
|:------|:---|:--------|
|observation|$n*k$-dimension array of ints|The actual number of observations for each type of merchant|
|alpha|k-dimension array of floats|Parameter of the distribution (k dimension for sample of dimension k)|
|iterations|int|The number of random draws from MCMC (default 500)|
|burn|int|The number of loops to warm-up for MCMC (default 100)|

**Output**: j-dimension array of floats: posterior calibarated postorior predictive p-value for each observation.
