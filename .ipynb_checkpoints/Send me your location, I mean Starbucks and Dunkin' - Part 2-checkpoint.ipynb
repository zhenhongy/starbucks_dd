## Starbucks liked a more 'diversified' neighborhood (Updated on October 24, 2017)

To find the key factors within a community for Starbucks and Dunkin', I decided to calculate the probability of each pair of merchants appearing in the same neiborhood. E.g. Suppose the sets of merchants around Store A, Store B and Store C are   
$$Set_A = \{banks\}, Set_B = \{banks, cafes\}, Set_C = \{banks, cafes, church\}$$
Then the probability matrix would be like this:

||bank|cafe|church|
|---|:------|:---|:--------|
|**bank**|1|2/3|1/3|
|**cafe**|2/3|2/3|1/3|
|**church**|1/3|1/3|1/3|

Let's have a glance of the probability matrix. The higher the probability is, the stronger the influence or the relation is. ( $P_{i,i}$ represents the influence of i-th type of merchant in dertermining sites being selected. $P_{i,j}$ represents the relation between i-th and j-th types of merchant. )

```python
merchants_similarity_s = pd.DataFrame( np.dot( nearby[nearby_type][slice_starbucks].applymap(lambda x: 1 if x > 0 else 0).T,
                                               nearby[nearby_type][slice_starbucks].applymap(lambda x: 1 if x > 0 else 0)) / sum(slice_starbucks),
                                       columns = nearby_type, index = nearby_type)
merchants_similarity_d = pd.DataFrame( np.dot( nearby[nearby_type][slice_dunkin].applymap(lambda x: 1 if x > 0 else 0).T,
                                               nearby[nearby_type][slice_dunkin].applymap(lambda x: 1 if x > 0 else 0)) / sum(slice_dunkin),
                                       columns = nearby_type, index = nearby_type)
merchants_similarity_s.iloc[1:10,1:10]
```

```python
def create_graph(weight_matrix, label, threshold = 0.6):
    weight_node = [(node, {'weight': weight_matrix.iloc[node, node], 'label': label[node]}) for node in range(len(label))]
    weight_edge = [(e_1, e_2, {'weight': weight_matrix.iloc[e_1, e_2], 'label':"{} & {}".format(label[e_1], label[e_2])}) \
                   for e_1, e_2 in zip(np.where(weight_matrix > threshold)[0], np.where(weight_matrix > threshold)[1]) \
                   if e_1 < e_2]
    G = nx.Graph()
    G.add_nodes_from(weight_node)
    G.add_edges_from(weight_edge)
    return G
```

```python
from bokeh.plotting import figure
from bokeh.models.graphs import from_networkx
from bokeh.models import (HoverTool, EdgesAndLinkedNodes, NodesAndLinkedEdges, MultiLine, Circle, LinearColorMapper, ColorBar, 
                          LabelSet, ColumnDataSource, CustomJS, Div)
from bokeh.palettes import brewer

def create_network_plot(networkx_object, title=None, layout = nx.spring_layout, div = False, label = True,
                        node_size = 'inflate', node_color = 'blue', edge_color = 'weight', mute_nonselect = True,
                        hover_active = True, OccurrenceRate = False, canvas_x_range = None, canvas_y_range = None,
                        min_weight = 0, max_weight = 1, palette = 'OrRd', colorbar_location = 'right', toolbar_location = 'above'):
    try:
        palette = brewer[palette][9]
    except KeyError:
        if type(palette) is str:
            print('No {} within brewer. \'OrRd\' is used.'.format(palette))
            palette = brewer['OrRd'][9]
    except TypeError:
            palette = palette
    
    mapper = LinearColorMapper(palette=palette[::-1], low=min_weight, high=max_weight)
    if colorbar_location is not None:
        color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
    
    graph = from_networkx(networkx_object, layout, scale=1)
    graph.inspection_policy = EdgesAndLinkedNodes()
    graph.selection_policy = NodesAndLinkedEdges()

    if type(node_size) is int:
        graph.node_renderer.data_source.data['size'] = [node_size] * len(networkx_object.nodes())
    elif node_size == r'inflate':
        graph.node_renderer.data_source.data['size'] = [len(networkx_object[node])*2 for node in networkx_object.nodes()]
    elif node_size == r'deflate':
        graph.node_renderer.data_source.data['size'] = [np.log(len(networkx_object[node])+0.01)+1 for node in networkx_object.nodes()]
    graph.node_renderer.data_source.data['label'] = [networkx_object.node[node]['label'] for node in networkx_object.nodes()]
    graph.node_renderer.data_source.data['weight'] = [networkx_object.node[node]['weight'] for node in networkx_object.nodes()]
    if type(node_color) is list:
        graph.node_renderer.data_source.data['color'] = node_color
    if node_color == r'weight':
        node_glyph = Circle(size = 'size', line_alpha = 0, fill_color= {'field': 'weight', 'transform': mapper}, fill_alpha = 0.5)
    elif type(node_color) is list:
        node_glyph = Circle(size = 'size', line_alpha = 0, fill_color= 'color', fill_alpha = 0.5)
    else:
        node_glyph = Circle(size = 'size', line_alpha = 0, fill_color= node_color, fill_alpha = 0.5)
    graph.node_renderer.glyph = node_glyph
    if mute_nonselect:
        node_nonselection_glyph = Circle(fill_color='grey', line_alpha = 0, fill_alpha = 0.5)
        graph.node_renderer.nonselection_glyph = node_nonselection_glyph
    
    graph.edge_renderer.data_source.data['label'] = [networkx_object[edge[0]][edge[1]]['label'] for edge in networkx_object.edges_iter()]
    graph.edge_renderer.data_source.data['node0_label'] = [networkx_object.node[edge[0]]['label'] for edge in networkx_object.edges_iter()]
    graph.edge_renderer.data_source.data['node1_label'] = [networkx_object.node[edge[1]]['label'] for edge in networkx_object.edges_iter()]
    graph.edge_renderer.data_source.data['weight'] = [networkx_object[edge[0]][edge[1]]['weight'] for edge in networkx_object.edges_iter()]
    graph.edge_renderer.data_source.data['node0_weight'] = [networkx_object.node[edge[0]]['weight'] for edge in networkx_object.edges_iter()]
    graph.edge_renderer.data_source.data['node1_weight'] = [networkx_object.node[edge[1]]['weight'] for edge in networkx_object.edges_iter()]
    if edge_color == r'weight':
        graph.edge_renderer.glyph = MultiLine(line_color={'field': 'weight', 'transform': mapper})
    else:
        graph.edge_renderer.glyph = MultiLine(line_color = edge_color, line_alpha = 0.5)
    graph.edge_renderer.nonselection_glyph = MultiLine(line_color='grey')
    
    layout_data = graph.layout_provider.graph_layout
    node_with_edge = [node for node, nbs in networkx_object.adjacency_iter() if len(nbs) > 0]
    label_source = ColumnDataSource(dict(x = [layout_data[node][0] for node in node_with_edge],
                                         y = [layout_data[node][1] for node in node_with_edge],
                                         name = [graph.node_renderer.data_source.data['label'][node] for node in node_with_edge]))
    labels = LabelSet(x='x', y='y', text='name', level='glyph', text_font_size = "10pt", 
                      x_offset=5, y_offset=5, source=label_source, render_mode='canvas',
                      background_fill_alpha = 0.5, background_fill_color = 'white')

    if canvas_x_range is None:
        canvas_x_range = (min(label_source.data['x']) - 0.03, max(label_source.data['x']) + 0.03)
    if canvas_y_range is None:
        canvas_y_range = (min(label_source.data['y']) - 0.03, max(label_source.data['y']) + 0.03)
    plot = figure(title=title, x_range=canvas_x_range, y_range=canvas_y_range, plot_width = 400, plot_height = 400,
                  tools="pan,wheel_zoom,tap,lasso_select,reset", toolbar_location = toolbar_location, active_drag="lasso_select")
    plot.axis.visible = False
    plot.grid.grid_line_color = None
    plot.outline_line_color = None
    
    hover = HoverTool(tooltips=[
                                ("Id", "@label"),
                                ("Relation", "@weight")
                                ], line_policy='nearest')
    if OccurrenceRate:
        hover.tooltips.append(("OccurrenceRate","(@node0_label:@node0_weight, @node1_label:@node1_weight)"))
    plot.add_tools(hover) 
    if not hover_active:
        plot.toolbar.active_inspect = None;

    plot.renderers.append(graph)
    if colorbar_location is not None:
        plot.add_layout(color_bar, colorbar_location)
    if label:
        plot.add_layout(labels)
    if div:
        div = Div(width = 300)
        graph.node_renderer.data_source.callback = CustomJS(args = dict(div=div), code = """
            var inx = cb_obj.get('selected')['1d'].indices;
            var label = cb_obj.get('data').label;
            var text = "You select "+ inx.length + " stores:\\n";

            for (i = 0; i < inx.length; i++) {
                var store = "<b>" + label[inx[i]] + "</b>\\n";
                text = text.concat(store);
            }
            div.text = text;
        """)
        plot = row(plot, div)
    
    return plot
```

## Dunkin' really loves transit station, car repair store and gas station

### The number of each type of merchant should follow Multinomial + Dirichlet distribution

If we evaluate preference using probability, then the number of each type is close to $N_{AllType}*P_{ThisType}$. Supposedly, the number of each type should follow [Multinomial distribution](https://en.wikipedia.org/wiki/Multinomial_distribution) with a prior [Dirichlet distribution](https://en.wikipedia.org/wiki/Dirichlet_distribution).  
$$ (N_{Type_A}, N_{Type_B},..., N_{Type_Z}) \sim Multinomial(N_{Type_{All}}, p_{Type_A}, p_{Type_B}... , p_{Type_Z}) $$
$$ (p_{Type_A}, p_{Type_B}, ... , p_{Type_Z}) \sim Dirichlet(\alpha_{Type_A}, \alpha_{Type_B}, ... , \alpha_{Type_Z}) $$    
The plots below describe the distribution of probability for each type of merchant. Most of them seem follow beta distribution. However, some plots (e.g. ATM, Bank, Cafe, gym, laundry and others) show obvious double peaks, which means inconsistent preferences might be possible within Starbucks and Dunkin'.

```python
nearby_prob = pd.DataFrame(nearby[nearby_type].values/ nearby[nearby_type].sum(axis=1).values.reshape(-1,1), columns = nearby_type)
f, axarr = plt.subplots(4, 5, figsize = (20,20))
for i in range(4):
    for j in range(5):
        if (i * 5 + j) == len(nearby_type):
            break
        s = sns.distplot(nearby_prob[nearby_type[i * 5 + j]][(nearby['type']=='starbucks')], color = 'green', ax = axarr[i,j], bins=10)
        s = sns.distplot(nearby_prob[nearby_type[i * 5 + j]][(nearby['type']=='dunkin')], color = 'red',ax = s, bins=10)
        s.set(xlim=(0,None))
plt.show()
```

To evaluate preference, I fit the prior distribution for Starbucks and Dunkin' seperately, and update posterior distribution as a part of [Emperical Bayes method](https://en.wikipedia.org/wiki/Empirical_Bayes_method). After that, calibrated [Postorior Predictive P-value](http://www.stat.columbia.edu/~gelman/research/published/A6n41.pdf) was used to find out specific stores which was obviously different from the group. It is alternative [chi-square test for independence](https://en.wikipedia.org/wiki/Chi-squared_test) in Bayesian Statistics.

```python
#Try to combine all regions
slice_starbucks = (nearby['type'] == 'starbucks')
slice_dunkin = (nearby['type'] == 'dunkin')
alpha_prior_starbucks = fit_mvpolya(nearby[nearby_type][slice_starbucks].values)['params']
alpha_prior_dunkin = fit_mvpolya(nearby[nearby_type][slice_dunkin].values)['params']
alpha_whole_starbucks = alpha_prior_starbucks + nearby[nearby_type][slice_starbucks].values.sum(axis=0)
alpha_whole_dunkin = alpha_prior_dunkin + nearby[nearby_type][slice_dunkin].values.sum(axis=0)
```

The plot of probability of each merchants from Bayesian model is pretty similar to the plot of the average number of each merchants in the previous part.    

However, Dunkin' has a much stronger preference for *transit station*, *car repair store*, *gas station* and *church* than Starbucks. Besides, Dunkin' is more likely to be in the area with *laundry store*, *liquor store*, *park* and *school*. In contrast, Starbucks has a relatively stronger preference with *bank*, *beauty salon* and *gym* than Dunkin'. It basically conforms to the different markets they focuses on (i.e, commercial and business districts for Starbucks; residential districts and places related to commuting, like transit station and high way, for Dunkin').

```python
#Sampling from dirichlet distribution
def dirichlet_rvs(alpha):
    y = np.array([])
    for a in alpha:
        y = np.append(y, np.random.gamma(a))
    return y / sum(y)

#Compare each distribution: Return estimation and CI of scaled difference (i.e. (p_x-p_y) / (p_x+p_y))
def comparison(alpha1, alpha2):
    diff = np.array([])
    for i in range(5000):
        p1 = dirichlet_rvs(alpha1)
        p2 = dirichlet_rvs(alpha2)
        diff = np.append(diff, (p1 - p2)/(p1 + p2))
    diff = diff.reshape(5000,-1).T
    estimator = np.percentile(diff, 50, axis = 1)
    ci_lower = np.percentile(diff, 2.5, axis = 1)
    ci_upper = np.percentile(diff, 97.5, axis = 1)
    return {'estimator': estimator, 'ci_lower': ci_lower, 'ci_upper': ci_upper}
```

```python
nearby_sum_bayes = pd.DataFrame({'merchant': np.append(nearby_type, nearby_type), 
                                 'prob': np.append(alpha_whole_dunkin/sum(alpha_whole_dunkin), alpha_whole_starbucks/sum(alpha_whole_starbucks)),
                                 'type': np.array(['dunkin']*len(nearby_type) + ['starbucks']*len(nearby_type))})

ci = comparison(alpha_whole_starbucks, alpha_whole_dunkin)
ci['ci_lower'] = ci['ci_lower'][:-1]
ci['ci_upper'] = ci['ci_upper'][:-1]
ci['estimator'] = ci['estimator'][:-1]
ci['merchant'] = nearby_type[:-1].tolist()
ci['starbucks'] = nearby_sum_bayes[nearby_sum_bayes['type'] == 'starbucks'].prob.values[:-1].tolist()
ci['dunkin'] = nearby_sum_bayes[nearby_sum_bayes['type'] == 'dunkin'].prob.values[:-1].tolist()
```

## Interestingly, Starbucks' preference could not describe its stores in major commercial area, while Dunkin's preference is less applicable in residential area

More than half of Starbucks stores in major commercial area and Allston are different from the average level. In contrast, regions, that are most distinctive to the average level, for Dunkin' are the places outside of major commercial area.

```python
import pymc3 as pm
#Discrepancy measurement for Postorior predictive p-value
def discrepancy(y, p):
    y_expectation = p * sum(y)
    return sum(np.array([(y1 - y2)**2 for y1, y2 in zip(y, y_expectation)]))
    
#Gibbs samppling for Dirichlet-Multinomial Distribution
def MCMC(iterations, burn, alpha, n):
    p = dirichlet_rvs(alpha)
    y = np.random.multinomial(n, p)
    p_simulation = []
    y_simulation = []
    for i in range(iterations+burn):
        alpha = alpha + y
        p = dirichlet_rvs(alpha)
        y = np.random.multinomial(n, p)
        if i >= burn:
            p_simulation.append(p)
            y_simulation.append(y)   
    return (p_simulation, y_simulation)

#Postorior predictive p-value
def ppp(observation, alpha, iteration = 500, burn = 100):
    p_simulation, y_replicate = MCMC(iteration, burn ,alpha , sum(observation))
    d_obs = []
    d_sim = []
    for p, y in zip(p_simulation, y_replicate):
        d_obs.append(discrepancy(observation, p))
        d_sim.append(discrepancy(y, p))
    return np.mean(np.array(d_sim) >= np.array(d_obs))

#Posterior Calibarated Postorior predictive p-value
def post_cppp(observation, alpha, iteration = 500, burn = 100):
    ppp_obs = np.array([ppp(observation, alpha)])
    
    ppp_sim = []
    p_simulation, y_replicate = MCMC(iteration, burn ,alpha , sum(observation))
    i = 1
    for y in y_replicate:
        ppp_sim.append(ppp(y,alpha))
        if i%100 == 0:
            print("{}% is completed".format(i / 10.0))
    
    return np.mean(np.array([ppp_obs]) >= np.array(ppp_sim))
```

```python
alpha_individual = nearby[nearby_type].values.astype(float)
alpha_individual[slice_starbucks] = alpha_prior_starbucks + alpha_individual[slice_starbucks]
alpha_individual[slice_dunkin] = alpha_prior_dunkin + alpha_individual[slice_dunkin]
results_all = []
for i in range(len(nearby)):
    print("Work on from row {}".format(i))
    results_all.append(post_cppp(nearby[nearby_type].values[i], alpha_individual[i]))
    print("{}% is done.".format((i+1)/len(nearby)*100))
    with open('./data/results_all.pkl','wb') as f:
        pickle.dump(results_all, f)
```
